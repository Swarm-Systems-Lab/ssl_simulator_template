# Simulation Tutorial
This notebook demonstrates how to set up simulations using user-defined controllers, robot models, and visualization tools.

## Setup and Imports

```python
# Import standard libraries
import os

# Third-Party Libraries
import numpy as np
import matplotlib.pyplot as plt

# Import the Swarm Systems Lab Simulator
from ssl_simulator import SimulationEngine, load_sim, add_src_to_path
from ssl_simulator.visualization import set_paper_parameters
set_paper_parameters(fontsize=12)

# -------------------------------------------------------------------------------------
# Import user-defined modules
add_src_to_path("", deep=0)

from src.controllers import ExampleController, ExampleControllerHL
from src.robot_models import ExampleRobot
from src.visualization import ExamplePlot

# -------------------------------------------------------------------------------------
# Output folder for simulation logs
OUTPUT_DIR = os.path.join("..", "output")
os.makedirs(OUTPUT_DIR, exist_ok=True)
SIMDATA_FILE = os.path.join(OUTPUT_DIR, "simulation_data.csv")
```

## Define Initial State and Controllers

```python
# Example initial state
N = 5
x0 = [np.random.random((N,2))*4] 

# Construct simulator
dt = 0.01
simulator = SimulationEngine(time_step=dt, log_filename=SIMDATA_FILE)

# Add robot model and controller
simulator.set_robot_model(ExampleRobot, x0)
simulator.add_controller("ctrl", ExampleController, 2)
simulator.add_controller("ctrl_hl", ExampleControllerHL, "ctrl", np.pi/2)

# Connect controller outputs to robot inputs
simulator.connect_controller_to_robot("ctrl", {"u":"u"})
```

## Run Simulation

```python
tf = 10.0
simulator.run(tf)
```

## Load and Plot Simulation Data

```python
simulation_data, settings = load_sim(SIMDATA_FILE)
plotter = ExamplePlot(simulation_data)
plotter.plot()
```
